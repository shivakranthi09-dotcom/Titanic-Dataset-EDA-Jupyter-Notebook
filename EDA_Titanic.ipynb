```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/New folder/Internship/train.csv")
df.head()
```

```python
#Dataset info
df.info()

#Summary statistics
df.describe()



```

```python
#Missing values
df.isnull().sum()
```

```python
#Categorical value counts
df['Embarked'].value_counts()
```

```python
#Categorical value counts
df['Sex'].value_counts()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

#Missing values
df.isnull().sum()
```

```python
#histogram (passanger age distribution)
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Boxplot((To detect outliers))
sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.xlabel("Fare")
plt.show()
```

```python
# Scatterplot( Age vs Fare)
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
#Countplot  Gender vs Survival
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Barplot  Survival Rate by Class
sns.barplot(x='Passanger class', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Class")
plt.ylabel("Survival Rate")
plt.show()
```
